```python

mylist = requests.get(RSS_URL['EdgePoint'],timeout=2.50).content
mylist = BeautifulSoup(mylist,features="xml")
tmylist = mylist.find_all("item")

pod_01 = pd.DataFrame(columns=['Name', 'Published', 'Episode','Media'])
for i in tmylist:
    
    author = i.find('author').text
    ddate = i.find('pubDate').text
    uurl = str(i.find('enclosure')).split('url=')[-1].split('/>')[0].replace('"','').split('.mp3')[0]+".mp3"
    ttile = i.find('title').text
    pod_01 = pod_01.append( {'Name' : author,
                             'Published': ddate,
                            'Episode' : ttile, 
                            'Media' : uurl},ignore_index=True)
    

#print(f'URL: {uurl}  \nTitle: {ttile} ')
pod_01.head()

#urllib.request.urlretrieve(uurl, 'mp3.mp3')
# TODO: CHANGE TO FUNCTION
```
